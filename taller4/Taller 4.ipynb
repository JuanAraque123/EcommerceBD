```python
import pandas as pd
import time
from datetime import datetime as dt

from google.cloud import bigtable
from google.cloud.bigtable import column_family, row_filters
```

```python
doc1=pd.read_csv("archives/Plant_1_Generation_Data.csv")
doc2=pd.read_csv("archives/Plant_1_Weather_Sensor_Data.csv")
doc3=pd.read_csv("archives/Plant_2_Generation_Data.csv")
doc4=pd.read_csv("archives/Plant_2_Weather_Sensor_Data.csv")
INSTANCE_ID = 'bigtable-1'
client = bigtable.Client.from_service_account_json('unbosque-service-account.json', admin = True)
print('Connecting to database instance')
instance = client.instance(INSTANCE_ID)
```

```python
doc2=doc2.to_dict('records')
TABLE_ID = 'timeonly1{}'.format(time.time())
print('Creating the {} table'.format(TABLE_ID))
table = instance.table(TABLE_ID)
if not table.exists():
    table.create()
    print("Table created sucssesfully")
else:
    print("ERROR: Table {} already exists".format(TABLE_ID))
    
print('Creating all family columns')

ambientTemp_cfambientTempID = 'ambient_temp'
ambientTemp_cf = table.column_family(ambientTemp_cfambientTempID)
ambientTemp_cf.create()
```

```python
dt = dt.utcnow()
rows = []

print('Writing times into the table timeonly1')
for doc in doc2:
    
    row_key = 'timeonly1#{}'.format(doc['DATE_TIME']).encode()
    row = table.direct_row(row_key)
    
    row.set_cell(ambientTemp_cfambientTempID, 'AMBIENT_TEMPERATURE'.encode(),str(doc['AMBIENT_TEMPERATURE']), timestamp = dt)
    
    rows.append(row)

table.mutate_rows(rows)
```

```python
dt = dt.utcnow()
rows = []

print('Writing times into the table time_key1')
for doc in doc1:
    
    row_key = 'timeonly1#{}'.format(doc['DATE_TIME']+':'+doc['SOURCE_KEY']).encode()
    row = table.direct_row(row_key)
    
    row.set_cell(totalYield_cfTYieldID, 'TOTAL_YIELD'.encode(),int(doc['TOTAL_YIELD']), timestamp = dt)
    row.set_cell(sumDirectAltern_cfDirectAndAlternID, 'sum_Direct_Altern'.encode(),int(doc['AC_POWER']+doc['DC_POWER']), timestamp = dt)
    
    rows.append(row)

table.mutate_rows(rows)
```
